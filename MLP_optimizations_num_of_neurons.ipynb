```python
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as  f
from sklearn.model_selection  import train_test_split
```

```python
data = pd.read_csv("Iris.csv")
data.tail()
```

```python
data = data.drop(columns= "Id")
```

```python
data['Species'] = data['Species'].replace('Iris-setosa',0.0)
data['Species'] = data['Species'].replace('Iris-versicolor',1.0)
data['Species'] = data['Species'].replace('Iris-virginica',2.0)
```

```python
data
```

```python
X = data.drop('Species',axis=1)
y = data['Species']
X
```

```python
# convert thesse to numpy array
X = X.values
y = y.values
```

```python
X.dtype
```

```python
X ,y
```

```python
X_train ,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
y_train
```

```python
#convert features to tensors
X_train = torch.FloatTensor(X_train)
X_test= torch.FloatTensor(X_test)
X_train
```

```python
# convert labels to long values
y_train = torch.LongTensor(y_train)
y_test = torch.LongTensor(y_test)
y_train
```

```python

class Model(nn.Module):

    def __init__(self,in_features = 4,h1 = 8 ,h2 =9 ,output = 3):
      super().__init__()
      #self.fc1=nn.Linear(in_features ,h1)
      #self.fc2=nn.Linear(h1,h2)
      #self.out=nn.Linear(h2,output)
      self.two_Layers = nn.Sequential(nn.Linear(in_features ,h1),
      nn.ReLU(),nn.Linear(h1 ,h2), nn.ReLU() ,nn.Linear(h2 ,output))

    #def forward(self , x):
      # return f.relu(self.out(f.relu(self.fc1(x))))

    def forward(self ,x):
      return self.two_Layers(x)

```

```python
model = Model()
```

```python
critersion = nn.CrossEntropyLoss()
```

```python

optimizar = torch.optim.Adam(model.parameters(),lr = 0.01)
y_train
```

```python
 y_pred = model.forward(X_train)
 y_pred
```

```python
epochs = 120
losses = []
for i in range(epochs) :
  y_pred = model.forward(X_train) # Get the prdiction
  loss = critersion( y_pred ,y_train) # clacul error between y_predict and y_test
  losses.append(loss.detach().numpy)
  # print result for each 10 iterations:
  if i % 10 == 0:
    print(f"epoch : {i} and loss : {loss}")
  optimizar.zero_grad()
  loss.backward()
  optimizar.step()


```
